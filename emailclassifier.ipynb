# Spam emails classifier

```python
#import required libraries
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import string
import re
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#import required data
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Email-spam-detection/master/emails.csv")
```

download stopwords and using string punctuation and stopwords remove all stop words

```python
nltk.download("stopwords")
```

```python
def process(text):
    nonpunc = [char for char in text if char not in string.punctuation]
    nonpunc = "".join(nonpunc)
    clean = [word for word in nonpunc.split() if word.lower() not in stopwords.words("english")]
    return clean
```

```python
data['text'].apply(process)
```

```python
data.head()
```

```python
data['text'] =data['text'].map(lambda x: re.sub("[^a-zA-Z]" , " " ,x))
```

```python
data
```

seperating the input and output

```python
X = data.iloc[: , 0]
y= data.iloc[: , -1]
```

```python
X
```

```python
y
```

using countvectorizer convert it into binary type

```python
cv = CountVectorizer(max_features = 9000)
```

```python
X = cv.fit_transform(X).toarray()
```

```python
X
```

```python
y =y.values
```

```python
cv.get_feature_names()
```

split the data and apply the suitable algorithm

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X  , y , test_size = 0.2 , random_state = 30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB().fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)*100
```
